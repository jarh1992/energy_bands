```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def varshni_rel(T):
    eg0 = 1.533 # GaAs T=0
    α = 5.5e-4 # eV/K
    β = 198 # K
    egT = eg0 - (α * T**2)/(β + T)
    return egT

T = np.arange(301)
Eg = np.array([varshni_rel(t) for t in T])
plt.plot(T, Eg, '--')
plt.show()
```

```python
def passler(T, x=0):
    eg0 = 1.5176
    α = 4.6e-4
    Θ = 203
    p = 2.85
    a = 1 + np.power((2 * T / Θ), p)
    egT = eg0 - (α * Θ / 2) * (np.power(a, (1 / p)) - 1)
    return egT

T = np.arange(301)
Eg = np.array([passler(t) for t in T])
plt.plot(T, Eg, '--')
plt.show()
print([(t, e) for t, e in zip(T, Eg)][-1])
```
